**Data Preproccesing and cleaning**

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns



print("Missing Values:")
print(df.isnull().sum())


df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
df['TotalCharges'].fillna(df['TotalCharges'].mean(), inplace=True)


def outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR


    plt.figure(figsize=(8, 4))
    plt.subplot(1, 2, 1)
    sns.boxplot(df[column])
    plt.title(f'{column} Before Removing Outliers')


    df = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

    plt.subplot(1, 2, 2)
    sns.boxplot(df[column])
    plt.title(f'{column} After Removing Outliers')
    plt.tight_layout()
    plt.show()

    return df


numcolumns = ['tenure', 'MonthlyCharges', 'TotalCharges']
for column in numcolumns:
    print(f"\n outliers in {column}")
    df = outliers(df, column)

print("\nShape after removing outliers:", df.shape)




bincolumns = ['gender', 'Partner', 'Dependents', 'PhoneService', 'PaperlessBilling', 'Churn']
l= LabelEncoder()
for column in bincolumns:
    df[column] = l.fit_transform(df[column])

categcolumns = ['InternetService', 'OnlineSecurity', 'OnlineBackup',
                      'DeviceProtection', 'TechSupport', 'StreamingTV',
                      'StreamingMovies', 'Contract', 'PaymentMethod', 'MultipleLines']
df = pd.get_dummies(df, columns=categcolumns)


scaler = StandardScaler()
df[numcolumns] = scaler.fit_transform(df[numcolumns])

print("\nFinal shape:", df.shape)
print("\nColumns after preprocessing:", df.columns.tolist())
```

**Building machine learning models **

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, StandardScaler
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split

X = df.drop('Churn', axis=1)
y = df['Churn']


smote = SMOTE(random_state=42)
X_balance, y_balance = smote.fit_resample(X, y)

print("Shape before SMOTE:", X.shape)
print("Distribution before SMOTE:\n", pd.Series(y).value_counts(normalize=True))
print("\nShape after SMOTE:", X_balance.shape)
print("Distribution after SMOTE:\n", pd.Series(y_balance).value_counts(normalize=True))


X_train, X_test, y_train, y_test = train_test_split(X_balance, y_balance, test_size=0.2, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report


randomf_model = RandomForestClassifier(n_estimators=100, random_state=42)
randomf_model.fit(X_train, y_train)
randomf_pred = randomf_model.predict(X_test)
print("Random Forest Results:")
print("Accuracy:", accuracy_score(y_test, randomf_pred))
print("\nDetailed Report:")
print(classification_report(y_test, randomf_pred))
```

```python
from sklearn.linear_model import LogisticRegression


logisticr_model = LogisticRegression(random_state=42, max_iter=1000)
logisticr_model.fit(X_train, y_train)
logisticr_pred = logisticr_model.predict(X_test)
print("Logistic Regression Results:")
print("Accuracy:", accuracy_score(y_test, logisticr_pred))
print("\nDetailed Report:")
print(classification_report(y_test, logisticr_pred))
```

```python
from sklearn.tree import DecisionTreeClassifier


decisiont_model = DecisionTreeClassifier(random_state=42)
decisiont_model.fit(X_train, y_train)
decisiont_pred = decisiont_model.predict(X_test)
print("Decision Tree Results:")
print("Accuracy:", accuracy_score(y_test, decisiont_pred))
print("\nDetailed Report:")
print(classification_report(y_test, decisiont_pred))
```

```python
from sklearn.svm import SVC


svm_model = SVC(kernel='rbf', random_state=42)
svm_model.fit(X_train, y_train)
svm_pred = svm_model.predict(X_test)
print("SVM Results:")
print("Accuracy:", accuracy_score(y_test, svm_pred))
print("\nDetailed Report:")
print(classification_report(y_test, svm_pred))
```

**Building deep learning models**

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.optimizers import Adam


simple_nn = Sequential([
    Dense(64, activation='relu', input_dim=X_train.shape[1]),
    Dropout(0.3),
    Dense(32, activation='relu'),
    Dense(16, activation='relu'),
    Dense(1, activation='sigmoid')
])


simple_nn.compile(
    optimizer=Adam(learning_rate=0.001),
    loss='binary_crossentropy',
    metrics=['accuracy']
)


history_simple = simple_nn.fit(
    X_train, y_train,
    batch_size=32,
    epochs=50,
    validation_split=0.2,
    verbose=1
)
y_pred_nn = (simple_nn.predict(X_test) > 0.5).astype(int)
nn_accuracy = accuracy_score(y_test, y_pred_nn)
print("\nSimple Neural Network Results:")
print(f"Accuracy: {nn_accuracy:.4f}")
print("\nClassification Report:")
print(classification_report(y_test, y_pred_nn))
```

```python
from tensorflow.keras.layers import SimpleRNN
from tensorflow.keras.optimizers import RMSprop


X_train_reshaped = X_train.values.reshape((X_train.shape[0], 1, X_train.shape[1]))
X_test_reshaped = X_test.values.reshape((X_test.shape[0], 1, X_test.shape[1]))
X_train_reshaped = X_train_reshaped.astype(np.float32)
X_test_reshaped = X_test_reshaped.astype(np.float32)
y_train = y_train.astype(np.float32)


rnn_model = Sequential([
    SimpleRNN(50, input_shape=(1, X_train.shape[1])),
    Dense(25, activation='relu'),
    Dropout(0.2),
    Dense(1, activation='sigmoid')
])


rnn_model.compile(
    optimizer=RMSprop(learning_rate=0.005),
    loss='mean_squared_error',
    metrics=['accuracy']
)


history_rnn = rnn_model.fit(
    X_train_reshaped, y_train,
    batch_size=128,
    epochs=75,
    validation_split=0.2,
    verbose=1
)
y_pred_rnn = (rnn_model.predict(X_test_reshaped) > 0.5).astype(int)
rnn_accuracy = accuracy_score(y_test, y_pred_rnn)
print("\nRNN Results:")
print(f"Accuracy: {rnn_accuracy:.4f}")
print("\nClassification Report:")
print(classification_report(y_test, y_pred_rnn))
```

**Fine tuning to the best**

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import RandomizedSearchCV


param_dist = {
    'n_estimators': [200, 300, 400, 500],
    'max_depth': [20, 30, 40, 50, None],
    'min_samples_split': [2, 3, 4, 5],
    'min_samples_leaf': [1, 2, 3],
    'max_features': ['sqrt', 'log2'],
    'class_weight': ['balanced', 'balanced_subsample', None]
}


randomf = RandomForestClassifier(random_state=42)


randomf_random = RandomizedSearchCV( estimator=randomf, param_distributions=param_dist,n_iter=30,cv=3,scoring='accuracy',verbose=1,random_state=42,n_jobs=-1)


randomf_random.fit(X_train, y_train)


print("\nBest Parameters:")
print(randomf_random.best_params_)

best_randomf = randomf_random.best_estimator_
y_pred = best_randomf.predict(X_test)

print("\nResults Comparison:")
print(f"Original RF Accuracy: 85.41%")
print(f"Optimized RF Accuracy: {accuracy_score(y_test, y_pred)*100:.2f}%")
print("\nDetailed Report:")
print(classification_report(y_test, y_pred))
```
